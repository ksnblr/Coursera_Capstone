# Function to get 100 venues for a given latitude from Foursquare

```python
def get_100_venues(borough_latitude,borough_longitude):
    LIMIT = 100 # limit of number of venues returned by Foursquare API
    radius = 500 # define radius
    url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION, 
    borough_latitude, 
    borough_longitude, 
    radius, 
    LIMIT)
    return url # display URL
```

# URL Generator

```python
url =get_100_venues(43.7,-79.38)
```

# Creating functions to get the datatable of venues for a given latitude

```python
def get_category(row):
    
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
results = requests.get(url).json()
venues=results['response']['groups'][0]['items']
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
nearby_venues = json_normalize(venues) # flatten JSON
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]


#nearby_venues['venue.location.lat','venue.location.lng']
```

```python
nearby_venues['venue.categories'] = nearby_venues.apply(get_category, axis=1)
nearby_venues.columns = [ x.split('.')[-1] for x in nearby_venues.columns]
```

# Putting into loop for getting the venues for all latitude and longitude in the dataframe

```python
venues_list = []
for neighbour,lat,lon in zip(toronto_df.Neighbourhood,toronto_df.Latitude,toronto_df.Longitude):
    url =get_100_venues(lat,lon)
    results = requests.get(url).json()["response"]['groups'][0]['items']
    venues_list.append([(
        neighbour,
        lat,
        lon,
        v['venue']['name'], 
        v['venue']['location']['lat'], 
        v['venue']['location']['lng'],  
        v['venue']['categories'][0]['name']) for v in results])
    print ("Doing for {}".format (neighbour))
venue_df = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
venue_df.columns = ['Neighbourhood Name','Neighbour Lat','Neighbour Lon','Venue Name','Venue Lat','Venue Lon','Venue Cat']
venue_df.head()
```

# Venue hot encoding to list the table category wise

```python
venue_df_hot = pd.get_dummies(venue_df[['Venue Cat']], prefix="", prefix_sep="")
venue_df_hot['Neighbourhood Name'] = venue_df['Neighbourhood Name']
fixed_columns = [venue_df.columns[0]] + list(venue_df_hot.columns[:-1])
venue_df_hot = venue_df_hot[fixed_columns]
#venue_df_hot[venue_df_hot['Neighbourhood Name']=='Rouge, Malvern']
```

# Grouping all Venues based on Neighbourhood Names

```python
venue_df_hot_grouped = venue_df_hot.groupby('Neighbourhood Name').sum().reset_index()
venue_df_hot_grouped.head()
```

# Getting the top 10 categories from all Neighbourhood

```python
neighbour_list = venue_df_hot_grouped['Neighbourhood Name']
top_ten_list =[]
for neighbour in neighbour_list:

    temp = venue_df_hot_grouped[venue_df_hot_grouped['Neighbourhood Name']==neighbour].T.reset_index()
    temp.columns=['Neighbour','Frequency']
    temp = temp.iloc[1:]
    temp.sort_values('Frequency',ascending=False,inplace=True)
    top_ten_list.append(["{} ({})" .format(t[0],t[1]) for t in temp[:10].values])
df_top_ten = pd.DataFrame(top_ten_list)
df_top_ten.columns = ['#1 Favourite','#2 Favourite','#3 Favourite','#4 Favourite','#5 Favourite','#6 Favourite','#7 Favourite','#8 Favourite','#9Favourite','#10 Favourite',]
df_top_ten.set_index(neighbour_list,inplace=True)
df_top_ten.reset_index()
```

# Running the KClusters with 5 initially. Later, I used ELBOW curve to find the best K

```python
from sklearn.cluster import KMeans


# set number of clusters
kclusters = 5

venue_grouped_clustering = venue_df_hot_grouped.drop('Neighbourhood Name', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(venue_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

# Finding WCSS to get the best K

```python
import matplotlib.pyplot as plt
%matplotlib inline
wcss = []
cl_num = 15

for i in range(1,cl_num):
    kmeans = KMeans(i)
    kmeans.fit(venue_grouped_clustering)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```

# By looking at the plot, the best K would be 2 or 3

```python
plt.plot(range(1,15),wcss)
plt.show
plt.ylabel('wcss')
plt.xlabel('no of clusters')
```

# So, setting best k as 5

```python
# set number of clusters
kclusters = 5

venue_grouped_clustering = venue_df_hot_grouped.drop('Neighbourhood Name', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(venue_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

# Inserting the K values into the top 10 favourite table to see how the clusters are found

```python
df_top_ten.insert(0,'Cluster Label',kmeans.labels_)
df_top_ten.head()
```

```python
combined_toronto_df=toronto_df[['Neighbourhood','Latitude','Longitude']].set_index('Neighbourhood').join(df_top_ten)
combined_toronto_df.reset_index(inplace=True)
combined_toronto_df.head()
```

# Mapping

```python
# create map
import matplotlib.cm as cm
import matplotlib.colors as colors

latitude =43.7
longitude = -79.383
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(combined_toronto_df['Latitude'], combined_toronto_df['Longitude'], combined_toronto_df['Neighbourhood'], combined_toronto_df['Cluster Label']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
